importing packages

```python
import pandas as pd
```

importing data

```python
df = pd.read_csv('Mobile_Price_Classification-220531-204702.csv')
df.head()
```

identifying independent and dependent axis

```python
x = df.drop(['price_range'], axis = 1)
y = df['price_range']
```

```python
x.head()
```

```python
y.head()
```

Seperating train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y, test_size = 0.25, random_state = 2)
```

Importing dependencies for model

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

Building the model

```python
model = Sequential()
model.add(Dense(units = 8, activation = 'relu', input_dim = len(X_train.columns)))
model.add(Dense(units = 4, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd', metrics = 'accuracy')
```

Feeding data

```python
model.fit(X_train,Y_train, validation_data = (X_test,Y_test), epochs = 100, batch_size = 32)
```

Saving weights

```python
model.save('weights.h5')
```
